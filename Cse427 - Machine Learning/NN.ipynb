```python
import numpy as np
```

```python
#Input array
X=np.array([[1,0,1,0],[1,0,1,1],[0,1,0,1]])
#Output
y=np.array([[1],[1],[0]])
```

```python
print(X)
```

```python
print(y)
```

```python
#Sigmoid Function
def sigmoid (x):
    return 1/(1 + np.exp(-x))
```

```python
#Derivative of Sigmoid Function
def derivatives_sigmoid(x):
    return x * (1 - x)
```

```python
#Variable initialization
epoch=1000 #Setting training iterations
lr=0.1 #Setting learning rate
inputlayer_neurons = X.shape[1] #number of features in data set
hiddenlayer_neurons = 3 #number of hidden layers neurons
output_neurons = 1 #number of neurons at output layer
```

```python
print(inputlayer_neurons)
```

```python
#weight and bias initialization
wh=np.random.uniform(size=(inputlayer_neurons,hiddenlayer_neurons))
bh=np.random.uniform(size=(1,hiddenlayer_neurons))
wout=np.random.uniform(size=(hiddenlayer_neurons,output_neurons))
bout=np.random.uniform(size=(1,output_neurons))
```

```python
print(wh)
```

```python
print(bh)
```

```python
print(wout)
```

```python
print(bout)
```

```python
for i in range(epoch):

#Forward Propogation
    print("\n\n::::::::::FORWARD PROPAGATION:::::::::\n\n")
    hidden_layer_input1=np.dot(X,wh)
    print("H = X . Wh\n")
    print(hidden_layer_input1)
    hidden_layer_input=hidden_layer_input1 + bh
    print("\nH = X . Wh + bias_h")
    print(hidden_layer_input)
    hiddenlayer_activations = sigmoid(hidden_layer_input)
    print("\nH_result: After applying activation function")
    print(hiddenlayer_activations)
    output_layer_input1=np.dot(hiddenlayer_activations,wout)
    print("\nOutput = H_result . Wout")
    print(output_layer_input1)
    output_layer_input= output_layer_input1+ bout
    print("\nOutput = H_result . Wout + bias_out")
    print(output_layer_input)
    output = sigmoid(output_layer_input)
    print("\nCalculated result: After applying activation function\n")
    print(output)

#Backpropagation
    print("\n\n::::::::::BACK PROPAGATION::::::::::::::\n\n")
    E = y-output
    print("Error=target-calculated")
    print(E)
    slope_output_layer = derivatives_sigmoid(output)
    print("Derivative of sigmoid (output)")
    print(slope_output_layer)
    d_output = E * slope_output_layer
    print("Delta output sum")
    print(d_output)
    
    wout += hiddenlayer_activations.T.dot(d_output) *lr
    print("Updated weight h-->o")
    print(wout)
    #wout += (d_output/hiddenlayer_activations.T) *lr
    
    bout += np.sum(d_output, axis=0,keepdims=True) *lr
    print("Updated output bias")
    print(bout)
    
    slope_hidden_layer = derivatives_sigmoid(hiddenlayer_activations)
    
    Error_at_hidden_layer = d_output.dot(wout.T)
    
    
    d_hiddenlayer = Error_at_hidden_layer * slope_hidden_layer
    print("Delta hidden sum")
    print(d_hiddenlayer)
    
    
    wh += X.T.dot(d_hiddenlayer) *lr
    print("Updated weight i-->h")
    print(wh)
    
    bh += np.sum(d_hiddenlayer, axis=0,keepdims=True) *lr
    print("Updated bias i-->h")
    print(bh)
    
print("\nFinal result: After completing all iterations\n")
print(output)
```
